# Recession Prediction using Machine Learning

This is a machine learning model to predict recession within next 6 months in United states.
Factors considered - GDP, Unemployment rate, Comsumer price index(CPI), 3 months T-bill rates, NASDAQ index. All features are taken in quartely frequency.

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.svm import SVC
```

## Data Preprocessing
All the data from seperate csv's are merged into a single data frame. Labels are created using previous recession data as per NBER. If the current date in the dataset lies in the recession with 6 month period then it is labelled as 1 else 0.

```python
# reading seperate csv for all factors and converting it into qurterly frequency
gdp_df = pd.read_csv('./input/GDP.csv', parse_dates =["DATE"], index_col ="DATE")
gdp_df = gdp_df.resample('Q').mean()

unemp_df = pd.read_csv('./input/UNRATE.csv', parse_dates =["DATE"], index_col ="DATE")
unemp_df = unemp_df.resample('Q').mean()

dtb_df = pd.read_csv('./input/DTB3.csv', parse_dates =["DATE"], index_col ="DATE")
dtb_df = dtb_df.resample('Q').mean()

cpi_df = pd.read_csv('./input/CPI.csv', parse_dates =["DATE"], index_col ="DATE")
cpi_df = cpi_df.resample('Q').mean()

nasdaq_df = pd.read_csv('./input/NASDAQCOM.csv', parse_dates =["DATE"], index_col ="DATE")
nasdaq_df['NASDAQCOM_PCH'] = pd.to_numeric(nasdaq_df['NASDAQCOM_PCH'], errors='coerce')
nasdaq_df = nasdaq_df.resample('Q').mean()
```

```python
# makeing a common dataframe for all features
features = pd.DataFrame()
features['gdp']=gdp_df['A191RL1Q225SBEA']
features.set_index(gdp_df.index)
features['unrate']=unemp_df['UNRATE_PCH']
features['dtb3']=dtb_df['DTB3_CHG']
features['nasdaq']=nasdaq_df['NASDAQCOM_PCH']
features['cpi']=cpi_df['CPIAUCSL_PCH']
```

```python
# taking dataset from 1971
x_values=features['1971-09-30':]
```

```python
# labelling the dataset as 0 and 1 for non-recession and recession preiod respectively
NBER_recessions = { '1': {'Begin': '1957-08-01', 'End': '1958-04-01'},
                    '2': {'Begin': '1960-04-01', 'End': '1961-02-01'},
                    '3': {'Begin': '1969-12-01', 'End': '1970-11-01'},
                    '4': {'Begin': '1973-11-01', 'End': '1975-03-01'},
                    '5': {'Begin': '1980-01-01', 'End': '1980-07-01'},
                    '6': {'Begin': '1981-07-01', 'End': '1982-11-01'},
                    '7': {'Begin': '1990-07-01', 'End': '1991-03-01'},
                    '8': {'Begin': '2001-03-01', 'End': '2001-11-01'},
                    '9': {'Begin': '2007-12-01', 'End': '2009-06-01'}}
y_values = pd.DataFrame()
y_values['recession']=[0]*len(gdp_df)
y_values['recession6']=[0]*len(gdp_df)
for recession in NBER_recessions:
    end_condition = (NBER_recessions[recession]['End']
                    >= gdp_df.index)
    begin_condition = (gdp_df.index
                    >= NBER_recessions[recession]['Begin'])
    y_values.loc[end_condition & begin_condition, 'recession'] = 1
for i in range(len(y_values)):
    if(y_values['recession'][i]==1):
        y_values.loc[i : min(i + 2, len(y_values) - 1),
                                         'recession6'] = 1
y_values=y_values[2:]
```

## Model training and validation

Here we used support vector machine model for recession prediction. Since there is huge class imbalance, we cannot use simple cross-validation or train-test split. So, we used a different approach - 

- Splitted data into train and test sets using stratified k fold cross validation technique. We kept it 5-fold (k=5).
- Handled class imbalance in train set with one of three available methods: random undersampling, random oversampling or oversampling with SMOTE algorithm.
- Trained and tested the model, separately measuring the accuracy of determining a recession and accuracy of determining the absence of a recession.
- Repeated this multiple times and averaged the results.


```python
# stratified k-fold cross-validation
kf = StratifiedKFold(n_splits=5, random_state=42)

# lists fro different accuracy metrics
cross_val_f1_score_lst = []
cross_val_accuracy_lst = []
cross_val_recall_lst = []
cross_val_precision_lst = []

for train_index_ls, validation_index_ls in kf.split(x_values, y_values['recession6']):
    
    # keeping validation set apart and oversampling in each iteration using smote 
    train, validation = x_values.iloc[train_index_ls], x_values.iloc[validation_index_ls]
    target_train, target_val = y_values['recession6'].iloc[train_index_ls], y_values['recession6'].iloc[validation_index_ls]
    sm = SMOTE(random_state=42)
    X_train_res, y_train_res = sm.fit_sample(train, target_train)
    print (X_train_res.shape, y_train_res.shape)
    
    # training the model on oversampled 4 folds of training set
    svm = SVC(gamma='auto', probability=True)
    svm.fit(X_train_res, y_train_res)
    
    # testing on 1 fold of validation set
    validation_preds = svm.predict(validation)
    cross_val_recall_lst.append(recall_score(target_val, validation_preds))
    cross_val_accuracy_lst.append(accuracy_score(target_val, validation_preds))
    cross_val_precision_lst.append(precision_score(target_val, validation_preds))
    cross_val_f1_score_lst.append(f1_score(target_val, validation_preds))
    
print ('Cross validated accuracy: {}'.format(np.mean(cross_val_accuracy_lst)))
print ('Cross validated recall score: {}'.format(np.mean(cross_val_recall_lst)))
print ('Cross validated precision score: {}'.format(np.mean(cross_val_precision_lst)))
print ('Cross validated f1_score: {}'.format(np.mean(cross_val_f1_score_lst)))
```

```python
# overall prediction for plot usnig trained model
res=svm.predict_proba(x_values)
recession=res[:,1]
```
